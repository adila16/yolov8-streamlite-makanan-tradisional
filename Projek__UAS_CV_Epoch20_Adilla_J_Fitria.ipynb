```python
# Mengimport librrary yang dibutuhkan untk implementasi yolov8

from ultralytics import YOLO
import os
from PIL import Image
import matplotlib.pyplot as plt
import yaml
```

```python
# ====== LOAD DATA.YAML SECARA OTOMATIS ======
with open("/content/drive/MyDrive/POLITEKNIK CALTEX RIAU/SEMESTER 8/Computer vision/Projek UAS Computer Vision/dataset/data.yaml", "r") as f:
    data_yaml = yaml.safe_load(f)

class_names = data_yaml['names']
class_dict = {i: name for i, name in enumerate(class_names)}

print(f"📋 Kelas yang digunakan: {class_dict}")
```

```python
# ====== TRAINING MODEL YOLOv8 ======
model = YOLO("yolov8n.pt")  # kamu bisa pakai yolov8s.pt, m.pt, l.pt juga

model.train(
    data="/content/drive/MyDrive/POLITEKNIK CALTEX RIAU/SEMESTER 8/Computer vision/Projek UAS Computer Vision/dataset/data.yaml",
    epochs=20,
    imgsz=640,
    batch=16,
    name="yolov8_makanan"
)
```

```python
# ====== EVALUASI DI DATA VALIDASI (valid/images) ======
trained_model = YOLO("runs/detect/yolov8_makanan/weights/best.pt")

print("\n🧪 Mengevaluasi model di validasi...")
metrics = trained_model.val()

print("📈 Hasil evaluasi (validasi):")
print(metrics)

metrics = trained_model.val(plots=True)
```

## Tes Model Menggunakan Data TEST

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import os

print("\n🔍 Memprediksi gambar dari test/images...")

test_folder = "/content/drive/MyDrive/POLITEKNIK CALTEX RIAU/SEMESTER 8/Computer vision/Projek UAS Computer Vision/dataset/test/images"
image_files = [f for f in os.listdir(test_folder) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

for image_file in image_files:
    image_path = os.path.join(test_folder, image_file)
    results = trained_model.predict(image_path, save=False, conf=0.3)
    result = results[0]
    boxes = result.boxes
    image = Image.open(image_path)

    print(f"\n🖼️ Gambar: {image_file}")

    # Tampilkan gambar dan bounding box
    fig, ax = plt.subplots(1, figsize=(10, 8))
    ax.imshow(image)
    ax.axis('off')

    if boxes and len(boxes.cls) > 0:
        for box, cls_id, conf in zip(boxes.xyxy.tolist(), boxes.cls.tolist(), boxes.conf.tolist()):
            cls_id = int(cls_id)
            nama_makanan = class_dict.get(cls_id, 'Tidak ada informasi')
            x1, y1, x2, y2 = box

            # Gambar kotaknya
            rect = patches.Rectangle((x1, y1), x2 - x1, y2 - y1,
                                     linewidth=2, edgecolor='red', facecolor='none')
            ax.add_patch(rect)

            # Tampilkan label dan confidence di atas kotak
            label = f"{nama_makanan} ({conf:.2f})"
            ax.text(x1, y1 - 10, label, color='white', fontsize=10,
                    bbox=dict(facecolor='red', edgecolor='none', pad=2))
            print(f" → {nama_makanan} (Confidence: {conf:.2f})")
    else:
        print(" → Tidak ada informasi (tidak ada objek terdeteksi)")

    plt.show()
```

```python
# Buat file app.py
#Isinya kode yang ngedeteksi diatas
%%writefile app.py
import streamlit as st
from PIL import Image
from ultralytics import YOLO

# Load model
model = YOLO("best.pt")  # pastikan best.pt ada di direktori yang sama

st.title("🍽️ Deteksi Makanan Tradisional dengan YOLOv8")
st.write("Unggah gambar makanan, dan model akan mendeteksi jenis serta kalorinya.")

uploaded_file = st.file_uploader("Upload Gambar", type=["jpg", "jpeg", "png"])

if uploaded_file is not None:
    image = Image.open(uploaded_file)
    st.image(image, caption="Gambar yang Diupload", use_column_width=True)

    with st.spinner("🔍 Mendeteksi..."):
        results = model.predict(image, conf=0.3)
        result = results[0]

        boxes = result.boxes
        if boxes and len(boxes.cls) > 0:
            st.success("✅ Objek terdeteksi:")
            for box, cls_id, conf in zip(boxes.xyxy.tolist(), boxes.cls.tolist(), boxes.conf.tolist()):
                nama_makanan = model.names[int(cls_id)]
                st.write(f"🍴 {nama_makanan} ({conf:.2f})")
        else:
            st.warning("⚠️ Tidak ada objek terdeteksi.")

        st.image(result.plot(), caption="Hasil Deteksi", use_column_width=True)
```
